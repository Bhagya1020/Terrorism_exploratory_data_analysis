#Data Analysis

```python
At1= sns.countplot(data=dataset,x='attacktype1')
for bars in At1.containers:
  At1.bar_label(bars)
```

#visualizing the data

##Year wise Attrack

```python
year_att= dataset.groupby(['iyear'],as_index=False)['attacktype1'].sum().sort_values(by='iyear',ascending=True)
year_att
```

```python
year_att.plot(kind="bar",figsize=(15,5),fontsize=13)
sns.barplot(x="iyear",y="attacktype1",data= year_att)
```

#no. of killed in each year

```python
year_kill= dataset.groupby(['iyear'],as_index=False)['nkill'].sum().sort_values(by='iyear',ascending=True)
year_kill
```

```python
year_kill.plot(kind="bar",figsize=(15,5),fontsize=13)
sns.barplot(x="iyear",y="nkill",data=dataset)
```

#Region wise attacks

```python
year_reg= pd.crosstab(dataset.iyear,dataset.region_txt)
year_reg
```

```python
region_df= pd.DataFrame(df['region_txt'].value_counts())
region_df
```

```python
plt.pie(region_df["region_txt"],labels= region_df.index,autopct='%0.2f%%',radius=1.6)
plt.show()
```

#number of kills in each region

```python
sns.scatterplot(x='nkill',y='region_txt',data=dataset,hue="iyear")
plt.show()
```

```python
sns.lineplot(x='nkill',y='imonth',data=dataset)
plt.show()
```

#Country wise Attack- Top 20

```python
country_att= dataset.groupby(['country_txt'],as_index=False)['attacktype1'].sum()[1:21]
country_att
```

```python
country_att.plot(kind="bar",figsize=(10,4))
plt.title('country wise attacks',fontsize=13)
plt.xlabel("country",fontsize=13)
plt.xticks(fontsize=12)
plt.ylabel("no of attacks",fontsize=13)
plt.show()
```

#number of Attacks by each attack type

```python
number_att= dataset.groupby(['attacktype1_txt'],as_index=False)['attacktype1'].sum().sort_values(by="attacktype1",ascending=True)
number_att
```

#Target type wise Attack

```python
target_att= dataset.groupby(['targtype1_txt'],as_index=False)['attacktype1'].sum().sort_values(by="attacktype1",ascending=False)
target_att
```
